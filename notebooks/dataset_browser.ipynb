# Dataset Ground Truth Browser

Use this notebook to inspect generated BlobNet dataset samples alongside their ground-truth heatmaps and atom coordinates.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl
```

### Choose A Dataset

```python
path = '../outputs/datasets/random_2026_06_24a/'
split = 'train/'
max_samples = 8

files = os.listdir(path+split)
files.sort()
if len(files) > max_samples:
    files = files[:max_samples]
```

```python
images = []
targets = []
for f in files:
    dset = np.load(path + split + f)
    image = dset['image']
    target = dset['target']
    images.append(image)
    targets.append(target)

fig, ax = plt.subplots(2, 4, figsize = (12, 6), sharex=True, sharey=True)

for i in range(4):
    ax[0, i].imshow(images[i], cmap='grey')
    ax[1, i].imshow(targets[i], cmap='jet')

for a in ax.flatten():
    a.axis('off')
fig.tight_layout()
```
